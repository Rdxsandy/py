```python
import numpy as np
import pandas as pd

country =['india','Pakistan','USA','Nepal','Shrilanka']
#  if i want to convert it into series
pd.Series(country)
# series has two things one is index which is given by pandas and secod is value which is given by us
# data type of series is object in 99% of cases it will be string means pandas treated string as object
```

```python
runs =[12,44,56,89,89,98,100]
pd.Series(runs)
```

```python
# we can give own index to the series

marks =[98,90,78,89,77]
subjects=["hindi","english","maths","science","History"]
pd.Series(marks,subjects)
pd.Series(marks,subjects,name="my_marks")
```

```python
# we can also create Series with Dictionary

marks_dict ={"hindi":98,"english":90,"maths":78,"science":89,"History":77}
pd.Series(marks_dict,name="my_marks")

```

```python
# Attributes of series object
# 1- size
# 2-dtype
# 3-name
# 4-isunique  if series contains repeated item then it will return false otherwise it will return true
# 5-index  fetch all the indcies together
# 6-values

money = [7,9,7,8,9,8,9]
money_series = pd.Series(money)
print(money_series)

# money_series.size=7
# money_series.dtype=dtype('int64')
# money_series.name='money'
# print(money_series.name)=money

# money_series.is_unique=false
money_series.index=['a','b','c','d','e','f','g']
print(money_series.index)
print(money_series)
print(money_series.values)


```
