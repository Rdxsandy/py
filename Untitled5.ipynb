```python
# attributes of numpy array
# ndim  tells the dimension of the array
# reshape  reshape the existing array for 3d (x,y,z) x tells us how  many 2d array and y,z are the dimensions of 2d array
# size  tells the size of the array
# dtype  tell the datatype of the object
# itemsize tells the item size of the array elements

a = np.arange(1,11)
print(a.ndim)
print(a)
print()
b = np.arange(0,40)
b.reshape(2,2,10)
print(b)
b.size
b.dtype


```

```python
# array operations
#  axis 1 = for row and zero for column

a = np.arange(1,9)
print(a)
print(a*9)
print(a>8)
# all these operation are valid in np array
```

```python
a = np.random.random((3,3))
print(a)
print()
a = a*100
print(a)
print()
np.round(a*100)
print(a)

print(np.min(a))
print(np.max(a))
print(np.sum(a))
print(np.prod(a))
print(np.max(a,axis=0))
print(np.max(a,axis=1))
```

```python
# do product of matrix
a = np.arange(1,9).reshape(2,4)
b = np.arange(1,9).reshape(4,2)
print(a)
print()
print(b)
print()
print(a.dot(b))
```
