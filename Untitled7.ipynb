```python
# 1. min() – Minimum value

# What it does:
# Returns the smallest value in each column (or row).

# df.min()


# Use case:
# Find the lowest marks, minimum salary, least sales, etc.

# 2. max() – Maximum value

# What it does:
# Returns the largest value in each column.

# df.max()


# Use case:
# Highest score, max profit, top temperature, etc.

# 3. mean() – Average

# What it does:
# Calculates the arithmetic average of values.

# df.mean()


# Formula:

# mean
# =
# sum of values
# number of values
# mean=
# number of values
# sum of values
# 	​


# Use case:
# Average marks, average income, average sales.

# 4. sum() – Total

# What it does:
# Returns the sum of values in each column.

# df.sum()


# Use case:
# Total revenue, total units sold, total marks.

# 5. count() – Number of non-null values

# What it does:
# Counts non-missing (non-NaN) values.

# df.count()


# Interview tip:
# count() ≠ total rows (it ignores NaN).

# 6. median() – Middle value

# What it does:
# Returns the middle value after sorting data.

# df.median()


# Use case:
# Better than mean when data has outliers.

# 7. mode() – Most frequent value

# What it does:
# Returns the value that appears most often.

# df.mode()


# Use case:
# Most common category, most repeated score.

# 8. std() – Standard Deviation

# What it does:
# Measures how spread out the data is.

# df.std()


# Use case:
# Risk analysis, variability in marks/sales.

# 9. var() – Variance

# What it does:
# Square of standard deviation.

# df.var()


# Use case:
# Statistical and ML preprocessing.

# 10. describe() – Summary statistics ⭐

# What it does:
# Gives a complete statistical summary.

# df.describe()


# Output includes:

# count

# mean

# std

# min

# 25%, 50%, 75% (quartiles)

# max

# Interview favorite question

# 11. Axis parameter (important!)
# df.mean(axis=0)  # column-wise (default)
# df.mean(axis=1)  # row-wise

# Example DataFrame
# import pandas as pd

# df = pd.DataFrame({
#     "Marks": [70, 80, 90, 85],
#     "Age": [20, 21, 22, 20]
# })

# Function	Result Meaning
# df.min()	Minimum Marks & Age
# df.max()	Maximum Marks & Age
# df.mean()	Average Marks & Age
# df.describe()	Full statistical summary
```
