```python
# important dataframe methods

import numpy as np
import pandas as pd

```

```python
# value counts is valid on both seris as well as dataframe

ipl=pd.read_csv('/content/ipl-matches.csv')
ipl.head(2)
```

```python
# find the player who won most of the award of the man of the match in finals and semifinals

mask = ~ipl['MatchNumber'].str.isdigit()
playoffs = ipl[mask]
player_counts = playoffs['Player_of_Match'].value_counts()

print(player_counts)


```

```python
dec=ipl['TossDecision'].value_counts()
print(dec)
dec.plot(kind='pie')
```

```python
# number of matches of every team
(ipl['Team1'].value_counts()+ipl['Team2'].value_counts()).sort_values(ascending=False)
# we can perform operation on two series irrespective of order of index if index values are matching
```

```python
# next function is sort_values applicable on both(seris and dataframe)
data = {
    "roll_no": [101, 102, 103, 104, 105],
    "marks": [85, 90, np.nan, 78, np.nan],
    "cgpa": [8.2, np.nan, 7.5, 8.0, 6.9]
}

df = pd.DataFrame(data)
print(df)
#  by default  the null values appears at the last but if we make position first then its wil show on top
```

```python
df.sort_values('marks')
```

```python
df.sort_values('cgpa', na_position='first')
# to change permanantly we can make true inplace parameter
#  we we want to sort based on multiple columns then..
#  dataframename(['a','b'..],[asc,des..])
# we can pass the list of columns as well as asending and decending list
```
