```python
import torch
import random
```

```python
num_samples = 1000

data = []
labels = []

for _ in range(num_samples):

    cga = round(random.uniform(5, 10), 2)
    iq = random.randint(90, 150)
    certs = random.randint(0, 8)

    # Simple selection rule
    score = 0

    if cga > 7.5:
        score += 1
    if iq > 120:
        score += 1
    if certs >= 3:
        score += 1

    selected = 1 if score >= 2 else 0

    data.append([cga, iq, certs])
    labels.append([selected])
```

```python
x = torch.tensor(data, dtype=torch.float32)
y = torch.tensor(labels, dtype=torch.float32)

# converting the list into pytorch tensor

# print(x)
# print(y)
```

```python
print(x.shape[0])
```

```python
train_size = int(0.8* x.shape[0])
print(train_size)
```

```python
x_train  = x[:train_size]
y_train = y[:train_size]

x_test = x[train_size:]
y_test = y[train_size:]

# print(x_train.shape)
# print(y_train.shape)
# print(x_test.shape)
# print(y_test.shape)
```

```python
indices = torch.randperm(x_train.size(0))
# print(indices)

x_train = x_train[indices]
y_train = y_train[indices]
# X_mean = x.mean(dim=0)
# X_std = x.std(dim=0)
# # Avoid divide-by-zero
# X_std[X_std == 0] = 1

# # Normalize
# x= (x - X_mean) / X_std
```

```python
import torch
import torch.nn as nn

# -------- MODEL --------
class StudentNN(nn.Module):
    def __init__(self):
        super(StudentNN, self).__init__()
        self.layer1 = nn.Linear(3, 6)
        self.layer2 = nn.Linear(6, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.layer1(x)
        x = self.layer2(x)
        x = self.sigmoid(x)
        return x


model = StudentNN()

# -------- LOSS + OPTIMIZER --------
criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)

# -------- TRAINING LOOP --------
epochs = 100

for epoch in range(epochs):

    # Forward
    outputs = model(x_train)
    loss = criterion(outputs, y_train)

    # Backward
    optimizer.zero_grad()
    loss.backward()

    # Update weights
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss = {loss.item():.4f}")
```

```python
model.eval()

with torch.no_grad():
    outputs = model(x_test)
    predicted = (outputs > 0.5).float()

    accuracy = (predicted == y_test).sum() / len(y_test)

print("Test Accuracy:", accuracy.item())
```
